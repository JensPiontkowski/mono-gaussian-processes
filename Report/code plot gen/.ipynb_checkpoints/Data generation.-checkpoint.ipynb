```python
avgs = np.zeros((41,4,2))

names = ['norm','uni','suppGP','suppGPlin']
expandednames = ['GP Regression','GP with m(x)=ax+b','GP with m(x)~GP(0,K(x,x\'))','GP with m(x)~GP(ax+b,K(x,x\'))']


err_values = np.zeros((2,4,13,100))
x_vals = np.array(range(3,40,3))
print(len(x_vals))
print(err_values.shape)
for mod in range(4):
    for tpt in range(len(x_vals)):
        for a in range(1,101):
            inpfile = open('../../RUN/err_%s/%s_%d_%d.txt'%(names[mod],names[mod],x_vals[tpt],a),'r')
            values = inpfile.read()
            arrr = values.splitlines()
            err_values[0,mod,tpt,a-1] = float(arrr[0])
            err_values[1,mod,tpt,a-1] = float(arrr[1])
            inpfile.close()
        
#         err_values[0,mod,tpt-1] = err_values[0,mod,tpt-1]/400.0
#         err_values[1,mod,tpt-1] = err_values[1,mod,tpt-1]/400.0
            

            
            
fig,ax = plt.subplots(2,1)
fig.set_size_inches(10,15)         
ax[0].set_title("Extrapolation test")
ax[1].set_title("Full test")



for i in range(4):
    ax[0].plot(x_vals,np.mean(err_values[0,i,:], axis=1),label=names[i])
    ax[1].plot(x_vals,np.mean(err_values[1,i,:], axis=1),label=names[i]+'_full')
    ax[0].legend()
    ax[1].legend()

plt.savefig("Error Curve.png")
plt.show()
```

## Generate data ##

```python
cnt = 60
x = np.linspace(-5,5,cnt)
np.save('x',x)
```

```python
t = -1*scipy.stats.t.logpdf(x,1)
t = preprocessing.scale(t)
np.save('f',t)
```

```python
y = t + np.random.normal(0,0.1,cnt)
np.save('y',y)
```

```python
fig = plt.figure(figsize=(9,6))
plt.plot(x,y,label='Data')
print(x[10]-x[0],x[59]-x[49])
plt.plot([x[0],x[10]],np.repeat(-2.6,2),'r|-')
plt.plot([x[49],x[59]],np.repeat(-2.6,2),'r|-',label='Extrapolation test region')
plt.plot([x[0],x[59]],np.repeat(-2.9,2),'m|-',label='General test region')
plt.plot([x[10],x[49]],np.repeat(-2.3,2),'g|-',label='Training data region')
plt.legend()
plt.title('Noisy minus log student\'s t-distrubution (%d data points)'%cnt)
plt.savefig('Input distribution')
```

# Real world data

```python
data = io.loadmat('./data1.mat')
```

```python
x = data['x'][0]
y = data['y'][0]
```

```python
x = x[np.logical_not(np.isnan(y))]
y = y[np.logical_not(np.isnan(y))]
```

```python
scaler = preprocessing.MinMaxScaler()
y = scaler.fit_transform(y[:,None]).flatten()

x = preprocessing.scale(x)
#y = preprocessing.scale(y)
```

```python
plt.plot(x,y,'r.')
#plt.savefig('realinput.png')
```

# Result

```python
nums = np.linspace(5,60,12,dtype=np.int)
print(nums)

errors = np.zeros((4,12),dtype=np.float32)
for model_idx in range(2,3):
    fig,ax = plt.subplots(4,3,sharex='all',sharey='all',figsize=(14,14))
    for plt_idx in range(12):
        errfile = open('./RUN/real4_%s/%s_%d.txt'%(names[model_idx],names[model_idx],nums[plt_idx]),'r')
        errors[model_idx,plt_idx] = float(errfile.read())#line().strip('\n')
        
        mean = np.load('./RUN/real4_%s/%s_%d_model.npy'%(names[model_idx],names[model_idx],nums[plt_idx]))
        q5 = np.load('./RUN/real4_%s/%s_%d_5q.npy'%(names[model_idx],names[model_idx],nums[plt_idx]))
        q95 = np.load('./RUN/real4_%s/%s_%d_95q.npy'%(names[model_idx],names[model_idx],nums[plt_idx]))
        params = np.load('./RUN/real4_%s/%s_%d_params.npy'%(names[model_idx],names[model_idx],nums[plt_idx]))
        a = int(plt_idx/3)
        b = plt_idx%3
        
        even = np.linspace(np.min(x),np.max(x),60)
        #print(a,b)
        ax[a,b].plot(x[idx[:20]],y[idx[:20]],'r.')
        ax[a,b].plot(x[idx[20:]],y[idx[20:]],'g+')
        ax[a,b].plot(even,mean[111:],'b')
        ax[a,b].text(0.8,1,'rho: %.3f'%(float(params[0])) )
        ax[a,b].fill_between(even,q5[111:],q95[111:],color="#dddddd")
        ax[a,b].set_title("no. of inducing points = %d"%(nums[plt_idx]))
        ax[a,b].plot(np.linspace(np.min(x),np.max(x),nums[plt_idx]),np.repeat(-0.5,nums[plt_idx]),'k|')
        
    fig.savefig('%s_inducing_vs_fit_20_points.png'%(names[model_idx]))
    
```
